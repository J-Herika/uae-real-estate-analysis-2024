```python
import pandas as pd
df = pd.read_csv("uae_real_estate_2024.csv")
```

```python
print(df.head())
print(df.info())
print(df.description)
```

```python
df.dropna(subset=['price','bedrooms','bathrooms'])
```

```python
avg_price_by_type = df.groupby('type')['price'].mean()
print(avg_price_by_type)
```

```python
verified_count = df['verified'].value_counts()
print(verified_count)
```

```python
df.to_csv("cleaned_uae_real_estate_2024.csv", index=False)
```
